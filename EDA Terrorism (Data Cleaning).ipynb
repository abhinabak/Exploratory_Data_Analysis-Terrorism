```python
df.rename(columns = {'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','provstate':'State','region_txt':'Region',
                     'city':'City','latitude':'Latitude','longitude':'Longitude','location':'Location','attacktype1_txt':'Attack Type',
                     'target1':'Target','nkill':'Killed','nwound':'Wounded','summary':'Summary','gname':'Group',
                     'targtype1_txt':'Target Type','weaptype1_txt':'Weapon Type','motive':'Motive'}, inplace = True)
```

```python
df.head()
```

```python
df1 = df[['Year','Month','Day','Country','State','Region','City','Latitude','Longitude','Attack Type',
          'Killed','Wounded','Target','Summary','Group','Target Type','Weapon Type','Motive']]
df1.head()
```

```python
df1.shape
```

```python
df1.isnull().sum()
```

```python
count = 0
for i in df1['Day']:
    if i == 0:
        count = count+1
    else:
        count = count
print('Number of Days entered as 0 is ',count)
```

```python
count1 = 0
for i in df1['Month']:
    if i == 0:
        count1 = count1+1
    else:
        count1 = count1
print('Number of Months entered as 0 is ',count1)
```

```python
df1.loc[df1['Day'] == 0, 'Day'] = df1.loc[df1['Day'] == 0, 'Day'].apply(lambda x: np.random.randint(1,32) if x == 0 else x)
df1.loc[df1['Month'] == 0, 'Month'] = df1.loc[df1['Month'] == 0, 'Month'].apply(lambda y: np.random.randint(1,13) if y == 0 else y)
```

```python
count = 0
for i in df1['Day']:
    if i == 0:
        count = count+1
    else:
        count = count
print('Number of Days entered as 0 is ',count)
```

```python
count1 = 0
for i in df1['Month']:
    if i == 0:
        count1 = count1+1
    else:
        count1 = count1
print('Number of Months entered as 0 is ',count1)
```

```python
df1.head()
```

```python
df1.describe(include = 'all')
```

```python
df1.loc[:,'Motive'] = df1['Motive'].fillna(value = 'NA')
df1.loc[:,'Summary'] = df1['Summary'].fillna(value = 'NA')
df1.loc[:,'City'] = df1['City'].fillna(value = 'NA')
df1.loc[:,'Target'] = df1['Target'].fillna(value = 'NA')

Killed_avg = df1['Killed'].mean(axis = 0)
Killed_avg = round(Killed_avg)
df1.loc[:,'Killed'] = df1['Killed'].fillna(value = Killed_avg)

Wounded_avg = df1['Wounded'].mean(axis = 0)
Wounded_avg = round(Wounded_avg)
df1.loc[:,'Wounded'] = df1['Wounded'].fillna(value = Wounded_avg)

df1.loc[:,'Latitude'] = df1['Latitude'].fillna(value = df1['Latitude'].median())
df1.loc[:,'Longitude'] = df1['Longitude'].fillna(value = df1['Longitude'].median())

df1.loc[:,'State'] = df1['State'].fillna(value = 'NA')
```

```python
df1.head()
```

```python
df1.isnull().sum()
```

```python
df1.loc[:,"Killed"] = df1["Killed"].fillna(0)
df1.loc[:,"Wounded"] = df1["Wounded"].fillna(0)
df1.loc[:,"Casualty"] = df1["Killed"]+df1["Wounded"]
```

```python
df1.to_csv('Cleaned_GlobalTerrorism.csv', index = False)
```
